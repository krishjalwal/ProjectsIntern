```python
import pandas
from pandas import DataFrame as df
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data
```

```python
data.describe()
```

```python
x = df(data,columns=['production_budget_usd'])
y = df(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('WorldWide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

Now For Linear Regression :

ho(x) = Oo + O1x choose Oo and O1 as y = c + mx
minimise : summation from i = 1 to n (yi - ho(xi))^2

This is called **Residual Sum of Squares**

```python
regression = LinearRegression()
regression.fit(x,y)
```

Slope Coefficient:

```python
regression.coef_  #theta_1
```

```python
regression.intercept_  #theta_2
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.plot(x,regression.predict(x),color='red',linewidth = 4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('WorldWide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

GoodNess of Fit : R^2
Now Predict revenue for a film with a $50 million budget.
theta1 and theta0 calculated above into the equation - theta0 + theta1x here x will be $50 million budget

real world data , how good our prediction

```python
regression.score(x,y)
```
